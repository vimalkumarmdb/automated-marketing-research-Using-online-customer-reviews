```python
from bs4 import BeautifulSoup
```

```python
import numpy as np
import matplotlib.pyplot as plt
from urllib.request import  urlopen
```

```python
import re
```

```python
url = "https://www.dpreview.com/reviews?category=cameras/"
```

```python
html =  urlopen(url)
```

```python
soup = BeautifulSoup(html)
```

```python
soup
```

```python
title = soup.title
print(title.text)
```

```python
links = soup.find_all('a',href = True)
```

```python
links
```

```python
link1 = []
for link in links:
    link1.append(link.get('href'))
```

```python
link1[1:5]
```

```python
links1=[]
for i in link1:
    matchObj = re.match( r'https://www.dpreview.com/reviews/',i, re.M|re.I)
    if(matchObj):
        links1.append(i)
```

```python
uniqueLinks=list(set(links1))
```

```python
len(uniqueLinks)
```

```python
uniqueLinks = uniqueLinks[0:8]
```

```python
def all_link(uniqueLinks):
    comments = []
    Brand = []
    for i in uniqueLinks:
        open_each_link = urlopen(i)
        html_soup = BeautifulSoup(open_each_link)
        Brand.append(html_soup.title.text)
        ac = html_soup.find_all('div',attrs={'class':'message middle'})
        for j in ac:
            comments.append(j.find_all('p')[0].text) 
        
    return comments,Brand
```

```python
Reviews, Brand = all_link(uniqueLinks)
```

```python
len(Reviews)
```

```python
Reviews
```

```python
Reviews[50]
```

```python
brands=[]
brand=['canon','sony','fuji','olympus','kodak','nokia','nikon','sandisk','epson','panasonic','samsung','sigma','google','tamrac','apple']
for i in uniqueLinks:
    for j in brand:
        if(i.find(j) != -1):
            brands.append(j)
```

```python
brands=list(set(brands))
```

```python
review_index={}
for b in brands:
    review_index[b]=0
for r in Reviews:
    rl=r.lower()
    for b in brands:
        if rl.find(b) != -1:
            review_index[b]+=1
```

```python
review_index
```

```python
brand_review={}
for b in brands:
    brand_review[b]=[]
for r in Reviews:
    rl=r.lower()
    for b in brands:
        if rl.find(b) != -1:
            brand_review[b].append(r)
```
